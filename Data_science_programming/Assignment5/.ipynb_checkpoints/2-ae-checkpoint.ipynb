# Autoencoder (AE)


"Autoencoding" (AE) is a data compression algorithm where the compression and decompression functions are 1) data-specific, 2) lossy, and 3) learned automatically from examples rather than engineered by a human. The compression and decompression functions are implemented with neural networks.

![autoencoder image](https://upload.wikimedia.org/wikipedia/commons/2/28/Autoencoder_structure.png)

#### Data-specific
An AE will only be able to compress data similar to what it has been trained on. This is different from, general-purpose audio or image compression algorithms.

#### Lossy
The decompressed outputs will be degraded compared to the original inputs.

#### Automatic Learning
It is easy to train specialized instances of the algorithm that will perform well on a specific type of input. It doesn't require any new engineering, just appropriate training data.

AEs are a *self-supervised* technique, a specific instance of supervised learning where the targets are generated from the input data. 

To build an autoencoder, one needs three things: an encoding function, a decoding function, and a distance function representing the amount of information loss between the decompressed representation of your data and the uncompressed representation (i.e. a "loss" function). 

Two interesting practical applications of autoencoders are *data denoising*, and *dimensionality reduction* for data visualization. With appropriate dimensionality and sparsity constraints, autoencoders can learn data projections that are more interesting than PCA or other basic techniques.


The following was inspired by https://blog.keras.io/building-autoencoders-in-keras.html. 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import keras
keras.__version__
```

```python
from sklearn import datasets
from sklearn.model_selection import train_test_split

if 0:  # easy & fast
    digits = datasets.load_digits()
    X = digits.data / 16.  # scaling
    img_shape = (8,8)  # 8x8 gray pixels
    y = digits.target
else:  # hard & slow
    mnist = datasets.fetch_mldata("MNIST original")
    X = mnist.data / 255.  # scaling
    img_shape = (28, 28)  # 28x28 gray pixels
    y = mnist.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.1)
    
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
plt.matshow(X_train[0].reshape(img_shape), cmap=plt.cm.gray_r)
y_train[0]
```

## Shallow AE (single layer)

A shallow AE is built like this:

```python
from keras.layers import Input, Dense
from keras.models import Model

input_img = Input(shape=(X_train.shape[1],))
encoded = Dense(32,  # this is the encoding dimensionality, vary this
                activation='relu')(input_img)
decoded = Dense(X_train.shape[1], activation='sigmoid')(encoded)

# this model maps an input to its reconstruction
autoencoder = Model(input_img, decoded)

autoencoder.summary()
```

```python
autoencoder.compile(optimizer='adadelta', 
                    loss='binary_crossentropy',
                    metrics=['accuracy'])

history = autoencoder.fit(X_train, X_train,
                epochs=20,
                batch_size=512,
                shuffle=True,
                verbose=1,
                validation_data=(X_test, X_test))

pd.DataFrame(history.history).plot()
```

Let's use the AE to encode and decode some digits from the *test* set:

```python
def draw_digits(x, autoencoder):
    x_hat = autoencoder.predict(x)
    n = len(x)
    for i in range(n):
        ax = plt.subplot(2, n, i + 1)
        plt.imshow(x[i].reshape(img_shape), cmap=plt.cm.gray_r)  # original
        ax.set_xticks(())
        ax.set_yticks(())
        ax = plt.subplot(2, n, i + 1 + n)
        plt.imshow(x_hat[i].reshape(img_shape), cmap=plt.cm.gray_r)  # reconstruction
        ax.set_xticks(())
        ax.set_yticks(())

plt.figure(figsize=(16, 4))
draw_digits(X_test[:10], autoencoder)
```

## Deep AE (multi-layer)

```python
encoded = Dense(32, activation='relu')(input_img)
encoded = Dense(16, activation='relu')(encoded)

decoded = Dense(32, activation='relu')(encoded)
decoded = Dense(X_train.shape[1], activation='sigmoid')(decoded)

autoencoder = Model(input_img, decoded)
```

```python
autoencoder.compile(optimizer='adadelta', 
                    loss='binary_crossentropy',
                    metrics=['accuracy'])

history = autoencoder.fit(X_train, X_train,
                epochs=20,
                batch_size=512,
                shuffle=True,
                verbose=1,
                validation_data=(X_test, X_test))

pd.DataFrame(history.history).plot()
```

Accuracy is about the same, but this one is compressing to 16 dimensions instead of 32.

```python
plt.figure(figsize=(16, 4))
draw_digits(X_test[:10], autoencoder)
```

## Convolutional AE (CAE)

```python
X_train = np.reshape(X_train, (len(X_train), *img_shape, 1))
X_test  = np.reshape(X_test,  (len(X_test),  *img_shape, 1))
X_train.shape
```

```python
from keras.layers import Conv2D, MaxPooling2D, UpSampling2D

input_img = Input(shape=X_train.shape[1:])

x = Conv2D(32, (3, 3), activation='relu', padding='same')(input_img)
x = MaxPooling2D((2, 2), padding='same')(x)
x = Conv2D(32, (3, 3), activation='relu', padding='same')(x)
encoded = MaxPooling2D((2, 2), padding='same')(x)

x = Conv2D(32, (3, 3), activation='relu', padding='same')(encoded)
x = UpSampling2D((2, 2))(x)
x = Conv2D(32, (3, 3), activation='relu', padding='same')(x)
x = UpSampling2D((2, 2))(x)
decoded = Conv2D(1, (3, 3), activation='sigmoid', padding='same')(x)

autoencoder = Model(input_img, decoded)
```

```python
autoencoder.compile(optimizer='adadelta', 
                    loss='binary_crossentropy',
                    metrics=['accuracy'])

history = autoencoder.fit(X_train, X_train,
                epochs=20,
                batch_size=128,
                shuffle=True,
                verbose=1,
                validation_data=(X_test, X_test))

pd.DataFrame(history.history).plot()
```

```python
plt.figure(figsize=(16, 4))
draw_digits(X_test[:10], autoencoder)
```

## Appplication: de-noising

```python
scale = 0.3  # the amount of noise
X_train_noisy = X_train + np.random.normal(loc=0.0, scale=scale, size=X_train.shape) 
X_test_noisy = X_test + np.random.normal(loc=0.0, scale=scale, size=X_test.shape) 

X_train_noisy = np.clip(X_train_noisy, 0., 1.)
X_test_noisy = np.clip(X_test_noisy, 0., 1.)

plt.subplot(1,2,1)
plt.imshow(X_train[0].reshape(img_shape), cmap='gray_r')
plt.subplot(1,2,2)
plt.imshow(X_train_noisy[0].reshape(img_shape), cmap='gray_r');
```

```python
plt.figure(figsize=(16, 4))
draw_digits(X_test_noisy[:10], autoencoder)
```

There seems to be some robustness to noise already, but now let's train for it explicitly, using the CAE:

```python
autoencoder.reset_states()
```

```python
autoencoder.compile(optimizer='adadelta', 
                    loss='binary_crossentropy',
                    metrics=['accuracy'])
                    
history = autoencoder.fit(X_train_noisy, X_train,
                epochs=20,
                batch_size=128,
                shuffle=True,
                verbose=1,
                validation_data=(X_test_noisy, X_test))

pd.DataFrame(history.history).plot();
```

```python
plt.figure(figsize=(16, 4))
draw_digits(X_test_noisy[:10], autoencoder)
```
