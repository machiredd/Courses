# Deep Neural Network (DNN)

A Deep Neural Network (DNN) is an Artificial Neural Network (ANN) with *multiple* hidden layers of units between the input and output layers. Similar to shallow ANNs, DNNs can model complex non-linear relationships. The extra layers enable composition of features from lower layers, giving the potential of modeling complex data with fewer units than a similarly performing shallow network. A Multi-Layer Perceptron (MLP) is an example of deep learning, but deep learning also includes Auto-Encoders (AE), Convolutional Neural Networks (CNN), Long-Short-Term-Memory (LSTM), and many other concepts.

DNNs are prone to overfitting because of the added layers of abstraction, which allow them to model rare dependencies in the training data. Regularization methods need to be applied during training to help combat overfitting. A recent regularization is *dropout* regularization, in which some number of units are randomly omitted from the hidden layers during training.

```python
import numpy as np
np.random.seed(1337)  # for reproducibility
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import keras
keras.__version__
```

## Example: digit classification

```python
from sklearn import datasets

if 0:  # easy & fast
    digits = datasets.load_digits()
    X = digits.data / 16.  # scaling
    img_shape = (8,8)  # 8x8 gray pixels
    y = digits.target
else:  # hard & slow
    mnist = datasets.fetch_mldata("MNIST original")
    X = mnist.data / 255.  # scaling
    img_shape = (28, 28)  # 28x28 gray pixels
    y = mnist.target
```

```python
plt.matshow(X[0].reshape(img_shape), cmap='binary')
y[0]
```

Models train better if we use a "one-hot" representation:

```python
from sklearn.preprocessing import LabelBinarizer

lb = LabelBinarizer()
Y = lb.fit_transform(y)  # binarize labels in a one-vs-all fashion

y[0], Y[0]  # to show the equivalence
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=.1)

print(X_train.shape)
print(Y_train.shape)
```

### Densely-connected Network

Now we are ready to specify the architecture of the DNN by specifying the type of layer (e.g. `Dense`, `Convolution2D`), the activation function (e.g. `relu`, `sigmoid`), and any additional layers (e.g. `Dropout`, `Merge`):

```python
from keras.models import Sequential
from keras.layers.core import Dense, Dropout, Activation

model = Sequential()
model.add(Dense(512, input_dim=np.prod(img_shape)))
model.add(Activation('relu'))
model.add(Dropout(0.2))  # randomly zero a fraction of input units at each update during training

model.add(Dense(512))
model.add(Activation('relu'))
model.add(Dropout(0.2))

model.add(Dense(10))
model.add(Activation('softmax'))

model.summary()
```

```python
model.compile(loss='categorical_crossentropy',
              optimizer='rmsprop',
              metrics=['accuracy'])

history = model.fit(X_train, Y_train,
                    batch_size=256,
                    epochs=20,
                    verbose=1, 
                    validation_data=(X_test, Y_test))
```

```python
pd.DataFrame(history.history).plot()

score = model.evaluate(X_test, Y_test, verbose=0)
print('Test accuracy:', score[1])
```

A more sophisticated way of vizualization of training and graphs, use TensorFlow's (the default backend of keras) *TensorBoard*. Start it with: 

    tensorboard --logdir=/tmp/tb

```python
from keras.callbacks import TensorBoard

board = TensorBoard(log_dir='/tmp/tb')
```

Let's train a little more, this time with the `callback` keyword:

```python
model.fit(X_train, Y_train,
                    batch_size=256,
                    epochs=2,
                    verbose=1,
                    validation_data=(X_test, Y_test),
                    callbacks=[board]);
```

### 2-D Convolution Network

![](https://upload.wikimedia.org/wikipedia/commons/6/68/Conv_layer.png)

A Convolutional Neural Network (CNN) is a type of ANN in which the connectivity pattern between its neurons is inspired by the organization of the animal visual cortex; i.e. CNNs use convolution (https://en.wikipedia.org/wiki/Convolution) in place of matrix multiplication. They are ideal for processing image data. CNNs capitalize on (1) sparse interactions, by using *small* kernels, and (2) parameter sharing, by using global kernels.

A typical layer of a convolutional network consists of three stages (see ﬁgure 9.7):

1. A layer performs several *convolutions* in parallel to produce a set of linear activations.
2. Each linear activation is run through a nonlinear activation function, such as `relu` (*detectors*)
3. A *pooling* function replaces the output of the net at a certain location with a summary statistic of the nearby outputs, e.g. the `MaxPooling` operation results in the maximum output within a rectangular neighborhood.

```python
input_shape = (img_shape[0], img_shape[1], 1)  # using just one channel / grayscale
X = X.reshape(X.shape[0], *input_shape)  # a 2-D CNN expects the input in this shape
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=.1)

print(X_train.shape)
print(Y_train.shape)
```

```python
from keras.layers import Conv2D, MaxPooling2D, Flatten

kernels = 16  # number of convolutional filters to use
kernel_size = (3, 3)  # convolution kernel size
pool_size = (2, 2)  # size of pooling area for max pooling

model = Sequential()
model.add(Conv2D(kernels, kernel_size, padding='valid', input_shape=input_shape))
model.add(Activation('relu'))

model.add(Conv2D(kernels, kernel_size))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=pool_size))
model.add(Dropout(0.25))

model.add(Flatten())
model.add(Dense(128))
model.add(Activation('relu'))
model.add(Dropout(0.5))
model.add(Dense(Y_train.shape[1]))
model.add(Activation('softmax'))

model.summary()
```

Note the difference in the number of parameters!

```python
model.compile(loss='categorical_crossentropy', 
              optimizer='adadelta', 
              metrics=['accuracy'])

history = model.fit(X_train, Y_train, 
                    batch_size=512, 
                    epochs=20,
                    verbose=1, 
                    validation_data=(X_test, Y_test),
                    callbacks=[board])
```

```python
pd.DataFrame(history.history).plot()

score = model.evaluate(X_test, Y_test, verbose=0)
print('Test accuracy:', score[1])
```

#### Visualization: Weights
For example, displaying the weights (i.e. the kernels) for the first layer:

```python
W, B = model.layers[0].get_weights()
W.shape, B.shape
```

```python
plt.figure(figsize=(8,8))
for row in range(4):
    for col in range(4):
        i = row * 4 + col
        plt.subplot(4, 4, i + 1)
        plt.imshow(W[:,:,0,i], cmap='bwr', vmin=-.5, vmax=.5)  # divergent colormap
        plt.xticks([]); plt.yticks([])
```

#### Visualization: Activations
Let's pick the first test datum as an input:

```python
inp = X_test[0, np.newaxis]  # inp must be 2-D

# the original scalar was built using flattened data
plt.imshow(inp.reshape(img_shape), cmap='binary');
```

```python
lb.inverse_transform(Y_test[:1])  # print true class
```

Applying the input, and calculating the activations of each layer:

```python
import keras.backend as K  # need access to the inner workings of keras

def get_activations(model, inputs):
    funs = [K.function([model.input] + [K.learning_phase()], [layer.output]) for layer in model.layers]
    return [fun([inputs, 0.])[0] for fun in funs]

activation = get_activations(model, inp)
```

Displaying the initial 2-d activation layers:

```python
for layer in range(6):
    plt.figure(figsize=(8,8))
    for row in range(4):
        for col in range(4):
            i = row * 4 + col
            plt.subplot(4, 4, i + 1)
            plt.imshow(activation[layer][0,:,:,i], cmap='bwr', vmin=-2, vmax=2)
            plt.xticks([]); plt.yticks([])
    plt.figtext(0.5, 0.9, f'activation layer {layer}: {model.layers[layer].name}', ha='center')
```

And the final 1-d activation layers:

```python
for layer in range(6, 12):
    plt.figure(figsize=(16,4))
    plt.imshow(np.atleast_2d(activation[layer][0]), cmap='bwr', vmin=-2, vmax=2)
    plt.title(f'activation layer {layer}: {model.layers[layer].name}')
```
