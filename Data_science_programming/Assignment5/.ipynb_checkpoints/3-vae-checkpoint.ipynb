# Variational Autoencoder (VAE)
A VAE restricts the encoding representation of a standard AE to learn a continuous *latent (inferred) variable model*, specifically the parameters of a unit ($\mu$=0, $\sigma$=1) Gaussian distribution that models the data. Because of this representation, you can also generate new data samples.

Training now requires two loss functions, the *reconstruction* loss (the accurate are the data reconstructed), and the *latent* loss based on the *Kullback–Leibler* (KL, https://en.wikipedia.org/wiki/Kullback–Leibler_divergence) divergence between the learned latent distribution and the unit Gaussian; this acts as a regularizer (although not strictly necessary).

While the traditional AE encoder generates a vector of real numbers as its encoding, the VAE encoder generates *two* vectors, the mean and the standard deviation of the Gaussians from which we *sample* the latent vector that is then used as input for the decoder.

VAEs have already shown promise in generating many kinds of complicated data, including handwritten digits, faces, house numbers, and predicting the future from static images.

![](http://kvfrans.com/content/images/2016/08/vae.jpg)

The following was inspired by https://blog.keras.io/building-autoencoders-in-keras.html.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import keras
keras.__version__
```

```python
from keras.datasets import mnist

(X_train, y_train), (X_test, y_test) = mnist.load_data()

img_shape = X_train.shape[1:]

X_train = X_train / 255.
X_test = X_test / 255.
X_train = X_train.reshape((len(X_train), -1))
X_test = X_test.reshape((len(X_test), -1))

X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
from keras.layers import Input, Dense

batch_size = 100

input_dim = np.prod(img_shape)
intermediate_dim = 512  # only one hidden layer -> shallow
latent_dim = 2

epsilon_std = 0.01  
```

Let's create the encoding layers first:

```python
x = Input(batch_shape=(batch_size, input_dim))

h_encoder = Dense(intermediate_dim, activation='relu')(x)

z_mean = Dense(latent_dim, activation='linear')(h_encoder)
z_log_sigma = Dense(latent_dim, activation='linear')(h_encoder)  # notice this is a second *parallel* layer
```

The `Lambda` layer wraps arbitrary expression as a Layer object:

```python
from keras.layers import Lambda
from keras import backend as K

def sampling(args):  
    z_mean, z_log_sigma = args
    epsilon = K.random_normal(shape=(batch_size, latent_dim), 
                              mean=0.,
                              stddev=epsilon_std)
    return z_mean + K.exp(z_log_sigma) * epsilon  # reparametrization trick, the randomness is in epsilon

z = Lambda(sampling)([z_mean, z_log_sigma])  # this layer combines the two previous layers
```

Now the decoding layers:

```python
h_decoder = Dense(intermediate_dim, activation='relu')  # we will reuse this object
h_decoded = h_decoder(z)

x_decoder = Dense(input_dim, activation='sigmoid')  # we will reuse this object
x_decoded = x_decoder(h_decoded)
```

Now we can create 3 models:

1. an end-to-end autoencoder mapping inputs to reconstructions
1. an encoder mapping inputs to the latent space
1. a generator that can take points on the latent space and will output the corresponding reconstructed samples.

```python
from keras.models import Model

# encoder, from inputs to latent space, using the mean as its output
encoder = Model(x, z_mean)

# decoder/generator, from latent space to reconstructed inputs
decoder_input = Input(shape=(latent_dim,))
decoder = Model(decoder_input, x_decoder(h_decoder(decoder_input)))

# end-to-end autoencoder, for reconstruction and training
vae = Model(x, x_decoded)
vae.summary()
```

Now we define the custom loss function:

```python
from keras import objectives

def vae_loss(x, x_hat):
    recon_loss = objectives.binary_crossentropy(x, x_hat) 
    kl_loss = - 0.5 * K.mean(1 + z_log_sigma - K.square(z_mean) - K.exp(z_log_sigma), axis=-1)
    return recon_loss + kl_loss
```

```python
vae.compile(optimizer='rmsprop', loss=vae_loss, metrics=['accuracy'])
history = vae.fit(X_train, X_train,
                  shuffle=True,
                  epochs=50,  # 10 quick, 50 for better
                  batch_size=batch_size,
                  verbose=1,
                  validation_data=(X_test, X_test))
```

```python
import pandas as pd

pd.DataFrame(history.history).plot();
```

```python
def draw_digits(x, autoencoder):
    x_hat = autoencoder.predict(x, batch_size=batch_size)
    n = len(x) // 10
    for i in range(n):
        ax = plt.subplot(2, n, i + 1)
        plt.imshow(x[i].reshape(img_shape), cmap=plt.cm.gray_r)  # original
        ax.set_xticks(())
        ax.set_yticks(())
        ax = plt.subplot(2, n, i + 1 + n)
        plt.imshow(x_hat[i].reshape(img_shape), cmap=plt.cm.gray_r)  # reconstruction
        ax.set_xticks(())
        ax.set_yticks(())

plt.figure(figsize=(16, 4))
draw_digits(X_test[:100], vae)  # must feed batches of batch_size
```

The latent space in our example is 2D, thus allowing for vizualization:

```python
X_test_encoded = encoder.predict(X_test, batch_size=batch_size)
plt.figure(figsize=(16, 16))
plt.scatter(X_test_encoded[:, 0], X_test_encoded[:, 1], c=y_test, marker='.', cmap='jet', );
plt.colorbar();
```

Because VAEs are generative models, we can also use them to generate new data. For example, we can scan the latent plane, sampling points at regular intervals, and generating the corresponding digit for each of these points. This gives us a visualization of the latent manifold that "generates" the digits:

```python
n = 25  # n^2 digits in figure

xx = np.linspace(-15, 15, n) * epsilon_std 
yy = np.linspace(-15, 15, n) * epsilon_std

digit_size = img_shape[0]
digits = np.zeros((digit_size * n, digit_size * n))
for i, yi in enumerate(xx):
    for j, xi in enumerate(yy):
        z = np.array([[xi, yi]])
        x_decoded = decoder.predict(z)[0].reshape(img_shape)
        digits[i * digit_size: (i + 1) * digit_size,
               j * digit_size: (j + 1) * digit_size] = x_decoded

plt.figure(figsize=(16, 16))
plt.imshow(digits, cmap='gray_r', extent=[xx[0], xx[-1], yy[0], yy[-1]]);
```

You can expand on this example by making the VAE deep, and/or applying convoluational layers.
